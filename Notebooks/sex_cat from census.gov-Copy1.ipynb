## API for pulling Data for Not Insured Males and Females per county in TN

```python
### sex,"sex_cat",county data for TN from 2010-2019

list_year = []

for year in range(2010,2020) :
    endpoint = 'https://api.census.gov/data/timeseries/healthins/sahie?get=NAME,NUI_PT,SEX_DESC,SEXCAT&for=county:*&in=state:47&time=' + str(year)
    params ={}
    response = requests.get(endpoint, params = params)
   ## print(year)
    ##print(response.content)
    res=response.json()
    res=pd.DataFrame(data=res)
    list_year.append(res)
    
```
