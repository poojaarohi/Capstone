## Income distribution of the population not insured /county in TN

```python
### NUI,"poverty",county data for TN from 2010-2019

list_year = []

for year in range(2010,2020) :
    endpoint = 'https://api.census.gov/data/timeseries/healthins/sahie?get=NAME,NUI_PT,IPR_DESC,IPRCAT&for=county:*&in=state:47&time=' + str(year)
    params ={}
    response = requests.get(endpoint, params = params)
    print(year)
    print(response.content)
    res=response.json()
    res=pd.DataFrame(data=res)
    list_year.append(res)
    
```

```python
##print(list_year)
```

```python
poverty = pd.DataFrame()
for index in range(0,10):
    poverty =  pd.concat([poverty,list_year[index]],axis=0)
poverty
```
