```python
### NIC_PT","NUI_PT","state",county data for TN from 2010-2019

list_year = []

for year in range(2010,2020) :
    endpoint = 'https://api.census.gov/data/timeseries/healthins/sahie?get=NIC_PT,NUI_PT,NAME&for=county:*&in=state:47&time=' + str(year)
    params ={}
    response = requests.get(endpoint, params = params)
    res=response.json()
    res=pd.DataFrame(data=res)
    list_year.append(res)
    
```

```python
new_data = pd.DataFrame()

for index in range(0,10):
    new_data =  pd.concat([new_data,list_year[index]],axis=0)
new_data
```
